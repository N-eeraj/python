```python
def fn(string):
    s = ""
    for i in range (len(string) - 1):
        if string[i] != string[i+1] and string[i] != string[i-1]:
            s += string[i]
    try:
        if string[-1] != s[-1]:
            s += string[-1]
        else:
            s = s[:-1]
    except:
        s = string[-1]
    if s == string:
        return s
    return fn(s)

fn("azxxxzy")
```
