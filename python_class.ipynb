### Using __init__ function

```python
class Person:
    def __init__(self, fn, ln, a):
        self.f_name = fn
        self.l_name = ln
        self.age = a
```

```python
john = Person('John', 'Wayne', 25)
print('Name:', john.f_name+' '+john.l_name, '\nAge:', john.age)
```

```python
matt = Person('Matt', 'Moris', 34)
print('Name:', matt.f_name+' '+matt.l_name, '\nAge:', matt.age)
```

### Object Methods

```python
class Student:
    
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        if gender == 'Male':
            self.gender = 'boy'
        else:
            self.gender = 'girl'
            
    def introduce(self):
        print('My name is', self.name)
        print('I am a', self.gender)
        print('I am', self.age, 'years old')
```

```python
jim = Student('Jim', 7, 'Male')
jim.introduce()
```

```python
ann = Student('Ann', 12, 'Female')
ann.introduce()
```

### Modify object Properties

```python
print(jim.age)
```

```python
jim.age=8
print(jim.age)
```

## Inheritance

```python
class Person:
    
    def __init__(self, name, age):
        self.name=name
        self.age=age
    
    def introduction(self):
        print('My name is', self.name)
        print('I am', self.age, 'years old')
```

```python
class Student(Person):
    
    def __init__(self, name, age, course, batch):
        Person.__init__(self, name, age)
        self.course = course
        self.batch = batch

    
    def sIntro(self):
        self.introduction()
        print('I study', self.course, 'in the batch of', self.batch)
```

```python
class Professional(Person):
    
    def __init__(self, name, age, job, exp):
        Person.__init__(self,name, age)
        self.job = job
        self.exp = exp

    def pIntro(self):
        self.introduction()
        print('I work as a', self.job)
        if self.exp == 0:
            print('I am a fresher')
        else:
            print('I have', self.exp, 'years of experience')
```

```python
grayson = Student('Richard Grayson', 19, 'Engineering', 2021)
grayson.sIntro()
```

```python
wayne = Professional('Bruce Wayne', 33, 'Businessman', 15)
wayne.pIntro()
```

#### Another Example for Inheritance

```python
class Hero:
    
    def __init__(self, name, comic):
        self.name = name
        self.comic = comic
    
    def intro(self):
        print(self.name, 'is from', self.comic, 'comics')
```

```python
class Marvel(Hero):
    
    def __init__(self, name, race, group, comic='Marvel'):
        super().__init__(name, comic)
        self.race = race
        self.group = group

    def marvelIntro(self):
        self.intro()
        print(self.name, 'is a', self.race)
        print(self.name, 'is a member of', self.group)
```

```python
class DC(Hero):
    
    def __init__(self, name, team, master, family, comic = 'DC'):
        super().__init__(name, comic)
        self.team = team
        self.master = master
        self.family = family
        
    def dcIntro(self):
        self.intro()
        print(self.name, 'is a member of', self.team)
        if self.master == True:
            print(self.name, 'is the leader of', self.family)
        else:
            print(self.name, 'is a protégé in the', self.family)
```

```python
tony=Marvel('Ironman', 'human', 'Avengers')
tony.marvelIntro()
```

```python
logan=Marvel('Wolverine', 'mutant', 'X-Men')
logan.marvelIntro()
```

```python
bruce=DC('Batman', 'Justice League', True, 'Batfam')
bruce.dcIntro()
```

```python
barry=DC('The Flash', 'Justice League', True, 'Speedsters')
barry.dcIntro()
```

```python
grayson=DC('Nightwing', 'Titans', False, 'Batfam')
grayson.dcIntro()
```

```python
wally=DC('Kid Flash', 'Young Justice', False, 'Speedsters')
wally.dcIntro()
```
