### While loop

```python
i=0
while i<2:
    print(i)
    i +=1
```

```python
#break is used to exit a loop
i=0
while i<5:
    print(i)
    i+=1
    if i==3:
        break
```

```python
#continue is used to skip current itteration
i=0
while i<5:
    i+=1
    if i==3:
        continue
    print(i)
```

## Functions

```python
#function without argument & return value
def fn1():
    print('hello world')
fn1()
```

```python
#function without argument & with return value
def fn2():
    return 'hello world'
print(fn2())
```

```python
#function with argument & without return value
def fn3(x):
    print('arg:',x)
fn3(2)
```

```python
#function with argument & return value
def fn4(x):
    return x**2
print(fn4(5))
```

```python
#function with multiple arguments
def fn5(a,b):
    return a+b
print(fn5(3,7))
```

```python
#function with variable arguments
def fn6(*args):
    for i in args:
        print(i, i**2)
fn6(1,3,4,7,10)
fn6(2,4,6)
```

```python
#function with variable arguments & keywords (kwargs)
def fn7(**a):
    for i,j in a.items():
        print(i,':',j)
fn7(x=5,y=3)
fn7(x=2,y=0,z=7)
```

### Recursive Function

```python
#factorial

def fact(x):
    if x==1:
        return 1
    else:
        return x * fact(x-1)

print(fact(int(input('Enter a Number:'))))
```

### Pattern

```python
for i in range(1,6):
    print(i*'*')
```

## Lambda Functions

```python
lFn = lambda x: print('hello\n'*x)
lFn(3)
```

## Map Functions

```python
lst = [1,2,5,7,3,10]
print(list(map(lambda x: x**2, lst)))
```

## Filter Functions

```python
x=[1,2,3,4,5,6]
print(list(filter(lambda a: a%2==1, x)))
```

## Reduce Functions

```python
from functools import reduce
a = list(range(1,11))
print(reduce(lambda x,y:x+y, a))
```
