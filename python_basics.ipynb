## String & String Functions

```python
s='Python is cool'
print(s)
```

```python
print(s.upper()) #returns string with all upper-cased letters
print(s.lower()) #returns string with all lower-cased letters
print(s.capitalize()) #returns string with 1st letter of each word as upper-case
```

```python
print('hello'.isupper()) #checks if string is in upper case
print('hello'.islower()) #checks if string is in lower case
print('HellO'.isupper()) #checks if string is in upper case
print('HellO'.islower()) #checks if string is in lower case
```

```python
print('5'.isdigit()) #checks if string is a digit
print('x'.isdigit())
print('5.0'.isdecimal()) #checks if string is a decimal
print('10'.isdecimal())
print('5x'.isdecimal())
```

```python
print(s)
print(s.index('c')) #returns first position of argument in string
print(s.index('o'))
print(s.index('on'))
print(s.index('ol'))
```

```python
print(s.index('x')) #throws exception if not found
```

```python
print(s)
print(s.find('cool')) #returns index if argument is found in the string
print(s.find('Cool')) #if argument isn't found in the string -1 is returned
```

### String Slicing

```python
print(s)
print(s[0]) #returns character at index 0
print(s[:]) #returns string
print(s[0:6]) #returns characters from index 0 to 6(excluding s[6])
print(s[10:-1]) #returns substring from 10th position to last(excluding last)
```

```python
print(s[:6]) #returns substring from 1st to 6th character
print(s[10:]) #returns substring from 10th to last character(including last)
```

```python
print(s[::1]) #returns string
print(s[::-1]) #returns reverse of the string
print(s[::2]) #returns string skipping 2 characters
print(s[::-2]) #returns reverse of string skipping 2 characters
```
